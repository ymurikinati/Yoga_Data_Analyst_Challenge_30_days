# Greetings everyone, today marks Day 7 of our project. We possess sales data from a supermarket and our objective is to analyze it using a range of Python libraries.

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("supermarket_sales - Sheet1.csv")


data.head()
```

```python
## let us see the  all payments methods 
plt.figure(figsize=(15,8))
sns.countplot(data=data,x='Payment')
plt.plot()
```

```python
## cheaking which gender use  which payment method most ?
plt.figure(figsize=(15,6))
sns.countplot(data=data,x='Payment',hue='Gender')
plt.show()
```

```python
## cheaking product line  useing countplot
plt.figure(figsize=(15,6))
sns.countplot(data=data,x='Product line')
```

```python
## cheaking product line with hue Gender   useing countplot
plt.figure(figsize=(15,6))
sns.countplot(data=data,x='Product line',hue='Gender')
```

```python
#plooting boxplot
sns.boxplot(data=data,y='Total')
```

```python
sns.boxplot(data=data,x='Gender',y='Total')
```

```python
sns.boxplot(data=data,x='Payment',y='Total')
```

```python
sns.boxplot(data=data,x='Payment',y='Total',hue='Gender')
```

```python
# cheakng   the all rating counts 
plt.figure(figsize=(15,6))
sns.countplot(data=data,x='Rating')
plt.xticks(rotation=80)
plt.show()
```

```python
##Payment methods versus total gross income 
plt.figure(figsize=(12,6))
plt.bar(x=data["Payment"],height=data["gross income"]);
plt.xticks(rotation=70);
```

```python
sns.countplot(x = 'Product line', data = data, palette = "mako")
plt.xticks(rotation=80)
plt.show()
```

```python
sns.lmplot(data=data, y="Quantity", x="Total")
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=data, x="Date", y="Tax 5%")
plt.xticks(rotation=80)
plt.legend()
plt.show()
```
